```python
import bs4
import urllib.request

nateUrl ="https://news.nate.com/"
htmlObject = urllib.request.urlopen(nateUrl)
webPage = htmlObject.read()
bsObject = bs4.BeautifulSoup(webPage, "html.parser")
print(bsObject)
```

```python
tag_div_class_snbArea = bsObject.find('div',{'class' : 'snbArea'})
print(tag_div_class_snbArea)
```

```python
li_list = tag_div_class_snbArea.find_all('li')
print(li_list)
```

```python
for li in li_list:
    a_tag = li.find('a')
    print(a_tag)
    print(f"a_tag = {a_tag}\n a_tag['href'] = {a_tag['href']}\n")
    print(f"a_tag.text = {a_tag.text}")
```
