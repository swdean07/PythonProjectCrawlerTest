```python
import bs4
webPage = open(r'./Sample03.html', 'rt', encoding='utf-8').read()
bsObject2 = bs4.BeautifulSoup(webPage, "html.parser")
print(bsObject2)
```

```python
tag_id_myId1 = bsObject2.find('div', {'id': 'myId1'})
print(tag_id_myId1)
```

```python
tag_id_myClass1 = bsObject2.find('div', {'class': 'myClass1'})
print(tag_id_myClass1)
```

```python
tag_div_myClass1 = bsObject2.find_all('div', {'class': 'myClass1'})
print(tag_div_myClass1)
```

```python
tag_ul_myClass2 = bsObject2.find('ul', {'class': 'myClass2'})
print(tag_ul_myClass2)
```

```python
tag_ul_myClass2 = bsObject2.find_all('ul', {'class': 'myClass2'})
print(tag_ul_myClass2)
```

```python
tag_li_myClass3 = bsObject2.find('li', {'class': 'myClass3'})
print(tag_li_myClass3)
```

```python
tag_li_myClass3 = bsObject2.find_all('li', {'class': 'myClass3'})
print(tag_li_myClass3)
```

```python
a_list = bsObject2.find_all('a')
print(a_list)
```

```python
for aTag in a_list:
    print(aTag['href'])
```

```python
li_list_all = bsObject2.find_all('li')
for tag_li in li_list_all:
    print(tag_li.text)
```

```python
for aTag in a_list:
    print(aTag.text)
```
